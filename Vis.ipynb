```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_pickle('df_cleaned')
```

```python
df
```

```python
sns.histplot(df['salary_in_usd'],kde=True, bins=7)
plt.title('Salary distribution', fontdict={'fontsize': 16})
plt.show()
```

# Mean salary by experience_level

```python
mean_salary_by_experince = df.groupby('experience_level')['salary_in_usd'].mean().sort_values().reset_index()
```

```python
order_list = mean_salary_by_experince['experience_level'].tolist()
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(14,7))
plt.subplot(1,2,1)
ax = sns.barplot(data=mean_salary_by_experince, x='experience_level', y = 'salary_in_usd', palette='Set2')
ax.set_title('Mean salary by experience_level', fontdict={'fontsize':16})
plt.subplot(1,2,2)
ax1 = sns.violinplot(data=df, x='experience_level', y='salary_in_usd', palette='Set2', order=order_list)
ax1.set_title('Salary distribution by Experience level', fontdict={'fontsize':16})
plt.tight_layout()
plt.show()
```

# Mean salary by employment type

```python
mean_salary_by_emp_type = df.groupby('employment_type')['salary_in_usd'].mean().sort_values().reset_index()
```

```python
plt.figure(figsize=(10,7))
plt.subplot(1,2,1)
ax = sns.barplot(data=mean_salary_by_emp_type, x='employment_type', y='salary_in_usd', palette='Set2')
ax.set_title("Mean salary by employment type", fontdict={'fontsize':16})

plt.subplot(1,2,2)
ax1 = sns.boxplot(data=df, x='employment_type', y='salary_in_usd', hue='employment_type', palette='Set3')
ax1.set_title("salary distribution by employment type", fontdict={'fontsize':16})
plt.tight_layout()
```

# salary distribution by company size

```python
mean_salary_by_company_size = df.groupby('company_size')['salary_in_usd'].mean().sort_values().reset_index()
```

```python
plt.figure(figsize=(11,7))
plt.subplot(1,2,1)
ax = sns.barplot(data=mean_salary_by_company_size, x='company_size', y='salary_in_usd', palette='Set2')
ax.set_title("Mean salary by Company Size", fontdict={'fontsize':16})

plt.subplot(1,2,2)
ax1 = sns.boxenplot(data=df, x='company_size', y='salary_in_usd', hue='company_size', palette='Set3')
ax1.set_title("salary distribution by company size", fontdict={'fontsize':16})
plt.tight_layout()
```

# salary distribution by job type

```python
plt.figure(figsize=(14,7))

ax1 = sns.boxenplot(data=df, x='job_type', y='salary_in_usd', hue='company_size', palette='Set2')
ax1.set_title("salary distribution by job type", fontdict={'fontsize':16})
plt.tight_layout()
```

#  Job type count

```python
plt.figure(figsize=(12,5))
sns.set_palette('Set2')
plt.subplot(1,2,1)
ax = df['job_type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
ax.set_title('Job type', fontdict={'fontsize':16} )

plt.subplot(1,2,2)
ax1 = df['experience_level'].value_counts().plot(kind='pie', autopct='%1.1f%%')
ax1.set_title('Experience level', fontdict={'fontsize':16} )
plt.show()
```

# Top 10 Jobs by Avg Salary

```python
top_10_jobs_salary = df.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False).reset_index().nlargest(10, 'salary_in_usd')
top_10_jobs_openings = df['job_title'].value_counts().sort_values(ascending=False).reset_index().nlargest(10, 'count')
```

```python
plt.figure(figsize=(14,7))
sns.set_palette('Set2')
plt.subplot(1,2,1)

ax = sns.barplot(data=top_10_jobs_salary, x='salary_in_usd', y='job_title', hue='job_title')
ax.set_title('Top 10 Jobs by Avg Salary')

plt.subplot(1,2,2)
ax1 = sns.barplot(data=top_10_jobs_openings, x='count', y='job_title', hue='job_title')
ax1.set_title('Top 10 Jobs by Jobs Opening')

plt.tight_layout()
```

# Top 10 Countries by Avg Salary

```python
top_10_countries_by_salary = df.groupby('company_location')['salary_in_usd'].mean().sort_values(ascending=False).reset_index().nlargest(10, 'salary_in_usd')
top_10_countries_by_openings = df['company_location'].value_counts().sort_values(ascending=False).reset_index().nlargest(10, 'count')
```

```python
plt.figure(figsize=(14,7))
sns.set_palette('Set2')
plt.subplot(1,2,1)

ax = sns.barplot(data=top_10_countries_by_salary, x='salary_in_usd', y='company_location', hue='company_location')
ax.set_title('Top 10 Countries by Avg Salary')

plt.subplot(1,2,2)
ax1 = sns.barplot(data=top_10_countries_by_openings, x='count', y='company_location', hue='company_location')
ax1.set_title('Top 10 Countries by Jobs opportunities')

plt.tight_layout()
```

# Top 10 residence Countries by Avg Salary

```python
top_10_res_countries_by_salary = df.groupby('employee_residence')['salary_in_usd'].mean().sort_values(ascending=False).reset_index().nlargest(10, 'salary_in_usd')
top_10_res_countries_by_openings = df['employee_residence'].value_counts().sort_values(ascending=False).reset_index().nlargest(10, 'count')
```

```python
plt.figure(figsize=(14,7))
sns.set_palette('Set2')
plt.subplot(1,2,1)

ax = sns.barplot(data=top_10_res_countries_by_salary, x='salary_in_usd', y='employee_residence', hue='employee_residence')
ax.set_title('Top 10 Countries by Avg Salary')

plt.subplot(1,2,2)
ax1 = sns.barplot(data=top_10_res_countries_by_openings, x='count', y='employee_residence', hue='employee_residence')
ax1.set_title('Top 10 Countries having most DS employees')

plt.tight_layout()
```

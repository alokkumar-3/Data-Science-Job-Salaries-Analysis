# **Fixing country names**

```python
cc = country_converter.CountryConverter()
```

```python
df_cleaned['company_location'] = cc.convert(df['company_location'], to='name_short')
```

```python
df_cleaned['employee_residence'] = cc.convert(df_cleaned['employee_residence'], to='name_short')
```

```python
df_cleaned.head()
```

# **Mapping**

```python
df_cleaned['experience_level'] = df_cleaned['experience_level'].map({
    'MI' : 'Mid',
    'SE' : 'Senior',
    'EN': "Entry",
    'EX' : 'Executive'
})
```

```python
df_cleaned['remote_ratio']  = df_cleaned['remote_ratio'].map({
    0: 'Onsite',
    50 : 'Hybrid',
    100 : 'Remote'
})
```

```python
df_cleaned.rename(columns={'remote_ratio' : 'job_type'}, inplace=True)
```

```python
df_cleaned['company_size'] = df_cleaned['company_size'].map({'L':'Large','M':'Medium', 'S':"Small"})
```

```python
df_cleaned['employment_type'] = df_cleaned['employment_type'].map({'FT':'Full-time', 'PT': 'Part-time', 'CT':'Contract', 'FL': 'Freelance'})
```
